# EX1:
Write a python function, named time_to_array_addition
the function gets as input an array or a list. Check the type and adds 1 to each argument in the array or list and prints the time it took.

```python
def time_to_array_addition(a):
  """
  This function tests if a is a numpy array or list
  """
  start_time = time.perf_counter()
  if type(a) == type([]): # a is list
    for i in range(len(a)):
      a[i] += 1
  elif type(a) == type(np.array([0])): # a is a numpy array
      a += 1
  else: # a is nither
    print('error')

  end_time = time.perf_counter()
  print(f"Time taken: {end_time - start_time}")
  return a
```
