```python
def ex2(array):
  max_num = max(array)
  new_array = np.arange(0,max_num+1,3)
  return new_array

print(ex2(np.array([1,2,6,9,15])))
```

4. Write a python function that takes two arrays and checks if all the elements are equal



```python
def check_if_equal(array1, array2):
  result = np.array_equal(array1, array2)
  return result

check_if_equal(np.array([1,2,3]),np.array([1,2,4]))
```

5. Write a python function that takes two arrays and checks if they contain the same variables not neccesserly in the same order

```python
def contain_same_variables(array1, array2):
  set1 = set(array1)
  set2 = set(array2)
  set1 = sorted(set1)
  set2 = sorted(set2)
  if set1 == set2:
    print('The arrays contains the same variables')
  else:
    print('The arrays does not contain the same variables')

a = np.array([5,6,7,3,4,8])
b = np.array([3,4,5,6,7])
contain_same_variables(a,b)
```
