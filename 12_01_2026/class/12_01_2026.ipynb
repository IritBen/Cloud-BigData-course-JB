write a function named ex2 that takes an array finds it's maximum value and than create a new array from zero to maximum value in steps of 3


```python
def ex2(array):
  max_num = max(array)
  new_array = np.arange(0,max_num+1,3)
  return new_array

print(ex2(np.array([1,2,6,9,15])))
```
